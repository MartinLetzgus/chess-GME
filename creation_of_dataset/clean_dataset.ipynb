```python
import pickle
import pandas as pd
```

```python
games1 = pickle.load( open( "evals1.p", "rb" ) )
games2 = pickle.load( open( "evals2.p", "rb" ) )
games3 = pickle.load( open( "evals3.p", "rb" ) )
games4 = pickle.load( open( "evals4.p", "rb" ) )
```

```python
df1 = pd.DataFrame(games1)
df2 = pd.DataFrame(games2)
df3 = pd.DataFrame(games3)
df4 = pd.DataFrame(games4)
```

```python
dfs = [df1,df2,df3,df4]
```

```python
df = pd.concat(dfs)
df.rename(columns={0: "url", 1: "eval"}, inplace=True)
```

```python
ma = pickle.load( open( "martinovichev.p", "rb" ) )
```

```python
ma = pd.DataFrame(ma)
```

```python
final= df.merge(ma)
```
